### Data cleaning 

```python
btc_data=bt_data[['btc_market_price','btc_market_cap','btc_n_transactions','btc_miners_revenue','btc_cost_per_transaction','btc_difficulty','btc_hash_rate','btc_cost_per_transaction_percent']]
```

```python
btc_data.head()
```

```python
btc_data.isnull().sum()
```

```python
btc_data['btc_difficulty'].describe()
```

```python
btc_data["btc_difficulty"].fillna("2.19", inplace = True)
```

```python
btc_data.isnull().sum()
```

###  Data modelling

```python
###split the data into target X and y
X = btc_data.drop('btc_market_price',axis=1).values
y = btc_data.btc_market_price.values
print(X.shape)
print(y.shape)
```

```python
##### scaling the data

from sklearn import preprocessing
```

```python
sX=preprocessing.scale(X)
```

```python
#### split the data into train and test
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(sX,y,test_size=0.3,random_state=12)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
btc_model=LinearRegression()
```

```python
btc_model.fit(X_train,y_train)
```

```python
### Make prediction
```

```python
y_pred=btc_model.predict(X_test)
```

```python
#### Model Evaluation
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_score=r2_score(y_test,y_pred)
MSE=mean_squared_error(y_test,y_pred)

print(r2_score)
print(MSE)
```
